# Chapitre 4 — Contrainte budgétaire

## Question du chapitre

Quels paniers le consommateur peut-il effectivement acheter ?

## Idée à comprendre d'abord

La contrainte budgétaire ne dit pas ce que le consommateur **veut**. Elle dit ce qu'il **peut** faire. Toute la suite du cours repose sur cette rencontre entre préférences et possibilités d'achat.

## 4.1. Équation de budget

Si le revenu est $R$ et les prix sont $p_1$ et $p_2$, alors :

$$
p_1x_1+p_2x_2 \leq R
$$

La frontière des paniers juste accessibles est :

$$
p_1x_1+p_2x_2=R
$$

## 4.2. Coordonnées à l'origine et pente

Si tout le revenu est consacré au bien 1 :

$$
x_1=\frac{R}{p_1}
$$

Si tout le revenu est consacré au bien 2 :

$$
x_2=\frac{R}{p_2}
$$

En isolant $x_2$, on obtient :

$$
x_2=\frac{R}{p_2}-\frac{p_1}{p_2}x_1
$$

La pente vaut donc :

$$
-\frac{p_1}{p_2}
$$

**Lecture analytique** : la pente de la droite de budget est $-p_1/p_2$ ; elle mesure la quantité de bien 2 à laquelle il faut renoncer pour obtenir une unité supplémentaire de bien 1.

**Lecture économique** : $p_1/p_2$ est le prix relatif du bien 1 en unités de bien 2. C'est une donnée de marché.

**Numéraire.** Seuls les prix relatifs comptent. Multiplier tous les prix et le revenu par un même $\lambda > 0$ ne change pas l'ensemble budgétaire : $\lambda p_1 x_1 + \lambda p_2 x_2 \leq \lambda R$ est identique à la contrainte initiale. On peut donc toujours normaliser un prix à 1 — c'est le **numéraire**.

## 4.3. Comment la droite de budget se déplace

### 1. Hausse du revenu

Si $R$ augmente et que les prix ne changent pas, la droite se translate parallèlement vers le haut.

### 2. Hausse de $p_1$

Si $p_1$ augmente :

- l'abscisse à l'origine $R/p_1$ diminue ;
- l'ordonnée à l'origine $R/p_2$ ne change pas ;
- la droite pivote vers l'intérieur autour de l'ordonnée à l'origine.

### 3. Hausse de $p_2$

Le raisonnement est symétrique : cette fois, c'est l'ordonnée à l'origine qui diminue et on pivote autour de l'abscisse à l'origine.

## 4.4. Ce qu'il faut voir immédiatement sur un graphique

Quand tu regardes une droite de budget, repère dans cet ordre :

1. les deux coordonnées à l'origine ;
2. la pente ;
3. la cause d'un déplacement : revenu ou prix ;
4. le sens du mouvement : translation ou rotation.

## Erreur classique

Confondre une **translation** de la droite de budget avec une **rotation**. Le revenu modifie les deux coordonnées à l'origine à la fois ; la variation d'un seul prix fait pivoter la droite.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

FIG_DIR = Path("microeco_v5_figures")
FIG_DIR.mkdir(exist_ok=True)

plt.rcParams.update({
    "figure.dpi": 160,
    "savefig.dpi": 200,
    "font.size": 11
})

def save_fig(name):
    plt.tight_layout()
    plt.savefig(FIG_DIR / f"{name}.png", bbox_inches="tight")
    plt.savefig(FIG_DIR / f"{name}.svg", bbox_inches="tight")
    plt.show()
```

## Graphique — 5 scénarios de variation de la contrainte budgétaire

**Référence commune.** $R=12$, $p_1=2$, $p_2=3$ → abscisse à l'origine $(6)$, ordonnée à l'origine $(4)$, pente $-2/3$.

| Scénario | Pente | Abscisse à l'origine $R/p_1$ | Ordonnée à l'origine $R/p_2$ |
|---|---|---|---|
| 1 — $p_1\downarrow$, $p_2$ constante | s'aplatit ($-1/3$) | ↑ (12) | inchangée (4) |
| 2 — $p_1\downarrow$, $p_2\uparrow$ | s'aplatit encore ($-1/6$) | ↑ (12) | ↓ (2) |
| 3 — $p_1\uparrow$, $p_2\downarrow$ | se redresse ($-8/3$) | ↓ (3) | ↑ (8) |
| 4 — $p_1\uparrow$, $p_2\uparrow$ | ambiguë ($-3/4$) | ↓ (4) | ↓ (3) |
| 5 — $R\uparrow$, prix constants | inchangée ($-2/3$) | ↑ (9) | ↑ (6) |

**Zones colorées.** Vert = gain de pouvoir d'achat (la nouvelle droite est au-dessus : on peut se permettre plus de bien 2 pour tout panier donné) ; rouge = perte. L'aire colorée est une indication qualitative et n'a pas d'autre interprétation économique comme elle peut en avoir dans d'autres graphiques (surplus du consommateur sur un graphique offre-demande, perte sèche d'une taxe, etc.).

**Scénarios 2 et 3.** Les deux droites se croisent à l'intérieur du plan : l'effet net dépend du panier consommé — gain sur un bien, perte sur l'autre.


```python
# 5 scénarios de variation de la contrainte budgétaire
# Référence : R=12, p1=2, p2=3  →  intercepts (6, 4), pente -2/3

R0, p10, p20 = 12, 2, 3

scenarios = [
    # (p1_new, p2_new, R_new, titre, note)
    (1, 3, 12,
     r"Scénario 1 : $p_1\downarrow$, $p_2$ constante",
     r"Pente $-1/3$   |   $R/p_1\uparrow$ (6→12)   |   $R/p_2$ inchangé (4)"),
    (1, 6, 12,
     r"Scénario 2 : $p_1\downarrow$, $p_2\uparrow$",
     "Pente $-1/6$   |   $R/p_1\\uparrow$ (6→12)   |   $R/p_2\\downarrow$ (4→2)\n"
     "Les droites se croisent → gain/perte selon le panier"),
    (4, 1.5, 12,
     r"Scénario 3 : $p_1\uparrow$, $p_2\downarrow$",
     "Pente $-8/3$   |   $R/p_1\\downarrow$ (6→3)   |   $R/p_2\\uparrow$ (4→8)\n"
     "Les droites se croisent → gain/perte selon le panier"),
    (3, 4, 12,
     r"Scénario 4 : $p_1\uparrow$, $p_2\uparrow$",
     r"Pente $-3/4$ (ambiguë)   |   $R/p_1\downarrow$ (6→4)   |   $R/p_2\downarrow$ (4→3)"),
    (2, 3, 18,
     r"Scénario 5 : $R\uparrow$, prix constants",
     r"Pente $-2/3$ inchangée   |   $R/p_1\uparrow$ (6→9)   |   $R/p_2\uparrow$ (4→6)"),
]

fig, axes = plt.subplots(2, 3, figsize=(15, 9))
axes_flat = axes.flatten()

for i, (p1n, p2n, Rn, title, note) in enumerate(scenarios):
    ax = axes_flat[i]

    x_end = max(R0 / p10, Rn / p1n) * 1.08
    x = np.linspace(0, x_end, 600)

    y_ref = np.maximum(R0 / p20 - (p10 / p20) * x, 0)
    y_new = np.maximum(Rn / p2n - (p1n / p2n) * x, 0)

    # Zones colorées gain / perte
    ax.fill_between(x, y_ref, y_new, where=(y_new >= y_ref),
                    color='tab:green', alpha=0.22, label='gain')
    ax.fill_between(x, y_new, y_ref, where=(y_ref > y_new),
                    color='tab:red', alpha=0.22, label='perte')

    ax.plot(x, y_ref, color='tab:blue',   lw=1.8, ls='--', label='référence')
    ax.plot(x, y_new, color='tab:orange', lw=2.0,          label='nouveau')

    # Intercepts
    for xv, yv, col in [
        (R0 / p10, 0, 'tab:blue'),   (Rn / p1n, 0, 'tab:orange'),
        (0, R0 / p20, 'tab:blue'),   (0, Rn / p2n, 'tab:orange'),
    ]:
        ax.scatter([xv], [yv], color=col, zorder=4, s=28)

    y_max    = max(R0 / p20, Rn / p2n) * 1.18
    x_max_pl = max(R0 / p10, Rn / p1n) * 1.1
    ax.set_xlim(0, x_max_pl)
    ax.set_ylim(0, y_max)
    ax.set_xlabel('$x_1$', fontsize=9)
    ax.set_ylabel('$x_2$', fontsize=9)
    ax.set_title(title, fontsize=9.5, pad=6)
    ax.text(0.5, -0.22, note, transform=ax.transAxes,
            ha='center', fontsize=8, style='italic')
    ax.legend(fontsize=7.5, loc='upper right')
    ax.grid(True, alpha=0.25)

axes_flat[5].set_visible(False)

fig.suptitle(
    '5 scénarios de variation de la contrainte budgétaire\n'
    r'(référence : $R=12$,  $p_1=2$,  $p_2=3$  →  intercepts $(6,\ 4)$, pente $-2/3$)',
    fontsize=11, fontweight='bold', y=1.01
)
plt.subplots_adjust(hspace=0.52, wspace=0.32)
save_fig('02b_5_scenarios_budget')
```

## Graphique - Contraintes budgétaires non-standard

Dans le cas standard, la contrainte est une droite. Trois dispositifs courants brisent cette linéarité.

### Rationnement

Une quantité maximale $\bar x_1$ est imposée sur le bien 1. La droite de budget est tronquée : elle suit la pente habituelle $-p_1/p_2$ jusqu'à $x_1 = \bar x_1$, puis devient un segment vertical (le consommateur ne peut pas acheter plus, quelle que soit sa capacité financière). L'ensemble budgétaire est strictement réduit par rapport au cas standard.

### Taxe ad valorem

Le prix effectif du bien 1 devient $p_1(1+t)$. La droite pivote vers l'intérieur autour du point d'ancrage $R/p_2$ (l'ordonnée à l'origine, inchangée), exactement comme une hausse de $p_1$. La recette fiscale est lisible sur le graphique comme l'écart horizontal entre la droite sans taxe et la droite avec taxe, évalué à la quantité choisie.

### Subvention plafonnée

Le prix effectif est $p_1 - s$ pour les $\bar x_1$ premières unités, puis $p_1$ au-delà. La droite est coudée en $x_1 = \bar x_1$ : plus plate (pente $-(p_1-s)/p_2$) jusqu'au seuil, puis elle reprend sa pente initiale $-p_1/p_2$ à partir du coude. C'est un exemple classique de contrainte budgétaire **non-convexe** : l'ensemble des paniers accessibles n'est pas convexe.


```python
# §4.5 — Contraintes budgétaires non-linéaires
# Référence commune : R=12, p1=2, p2=2  →  intercepts (6, 6), pente -1

R, p1, p2 = 12, 2, 2

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14, 5.5))

# ── 1. Rationnement ───────────────────────────────────────────────────────────
x_bar = 4

x_all = np.linspace(0, R / p1, 400)
y_all = R / p2 - (p1 / p2) * x_all

# Zone exclue : triangle entre x_bar et R/p1 sous la droite de budget
x_excl = np.linspace(x_bar, R / p1, 200)
y_excl = R / p2 - (p1 / p2) * x_excl
ax1.fill_between(x_excl, 0, y_excl,
                 color='tab:red', alpha=0.18, label='zone exclue')

ax1.plot(x_all, y_all, color='tab:blue', lw=1.5, ls='--', alpha=0.45,
         label='sans rationnement')

# Droite tronquée (segment + barre verticale)
x_rat = np.linspace(0, x_bar, 300)
ax1.plot(x_rat, R / p2 - (p1 / p2) * x_rat, color='tab:blue', lw=2)
y_kink1 = R / p2 - (p1 / p2) * x_bar
ax1.vlines(x=x_bar, ymin=0, ymax=y_kink1, color='tab:blue', lw=2)
ax1.axvline(x=x_bar, color='gray', lw=0.8, ls=':', zorder=0)

ax1.scatter([R / p1], [0],       color='tab:blue', alpha=0.4, zorder=4, s=30)
ax1.scatter([0],      [R / p2],  color='tab:blue', zorder=4, s=30)
ax1.scatter([x_bar],  [0],       color='tab:blue', zorder=4, s=30)
ax1.scatter([x_bar],  [y_kink1], color='tab:blue', zorder=4, s=30)

ax1.text(x_bar + 0.1, 0.22, r'$\bar x_1 = 4$', fontsize=9.5, color='gray')
ax1.set_xlim(0, 8); ax1.set_ylim(0, 8.5)
ax1.set_xlabel('$x_1$'); ax1.set_ylabel('$x_2$')
ax1.set_title(r'Rationnement ($\bar x_1 = 4$)', fontsize=10)
ax1.legend(fontsize=8, loc='upper right')
ax1.grid(True, alpha=0.25)

# ── 2. Taxe ad valorem ────────────────────────────────────────────────────────
t = 0.5                    # taux 50 %
p1_tax = p1 * (1 + t)     # prix effectif = 3

x_ref2 = np.linspace(0, R / p1,     300)
x_tax  = np.linspace(0, R / p1_tax, 300)

# Perte : toute l'aire entre les deux droites (avec taxe clampée à 0)
x_fill2    = np.linspace(0, R / p1, 300)
y_ref_fill = R / p2 - (p1     / p2) * x_fill2
y_tax_fill = np.maximum(R / p2 - (p1_tax / p2) * x_fill2, 0)
ax2.fill_between(x_fill2, y_tax_fill, y_ref_fill,
                 color='tab:red', alpha=0.2, label="perte de pouvoir d'achat")

ax2.plot(x_ref2, R / p2 - (p1     / p2) * x_ref2,
         color='tab:blue',   lw=1.8, ls='--', label='sans taxe')
ax2.plot(x_tax,  R / p2 - (p1_tax / p2) * x_tax,
         color='tab:orange', lw=2,
         label=f'avec taxe ($t = {int(t*100)}\\%$)')

# Point pivot sur l'axe vertical
ax2.scatter([0], [R / p2], color='gray', zorder=5, s=35)
ax2.text(0.15, R / p2 + 0.2, 'pivot', fontsize=8.5, color='gray')
ax2.scatter([R / p1, R / p1_tax], [0, 0],
            color=['tab:blue', 'tab:orange'], zorder=4, s=30)

ax2.set_xlim(0, 8); ax2.set_ylim(0, 8.5)
ax2.set_xlabel('$x_1$'); ax2.set_ylabel('$x_2$')
ax2.set_title(f'Taxe ad valorem ($t = {int(t*100)}\\%$, $p_1^{{eff}} = {p1_tax:.0f}$)',
              fontsize=10)
ax2.legend(fontsize=8, loc='upper right')
ax2.grid(True, alpha=0.25)

# ── 3. Subvention plafonnée ───────────────────────────────────────────────────
s_sub   = 1               # subvention unitaire
x_bar_s = 3               # seuil (unités subventionnées)
p1_s    = p1 - s_sub      # prix effectif subventionné = 1

# Coude en x1 = x_bar_s
x2_kink  = R / p2 - (p1_s / p2) * x_bar_s    # = 6 - 0.5*3 = 4.5
R_rest   = R - p1_s * x_bar_s                 # budget restant après seuil = 9
x1_max_s = x_bar_s + R_rest / p1              # = 3 + 4.5 = 7.5

x_seg1 = np.linspace(0,       x_bar_s,  200)
x_seg2 = np.linspace(x_bar_s, x1_max_s, 200)
y_seg1 = R / p2 - (p1_s / p2) * x_seg1
y_seg2 = x2_kink - (p1 / p2) * (x_seg2 - x_bar_s)

x_ref3 = np.linspace(0, R / p1, 300)
y_ref3 = R / p2 - (p1 / p2) * x_ref3

# Gain : toute l'aire où la droite subventionnée est au-dessus de la référence
x_full    = np.linspace(0, x1_max_s, 400)
y_ref_full = np.maximum(R / p2 - (p1 / p2) * x_full, 0)
y_sub_full = np.where(
    x_full <= x_bar_s,
    R / p2 - (p1_s / p2) * x_full,
    x2_kink - (p1 / p2) * (x_full - x_bar_s)
)
y_sub_full = np.maximum(y_sub_full, 0)
ax3.fill_between(x_full, y_ref_full, y_sub_full,
                 where=(y_sub_full > y_ref_full),
                 color='tab:green', alpha=0.22, label='gain (subvention)')

ax3.plot(x_ref3, y_ref3, color='tab:blue',  lw=1.8, ls='--', label='sans subvention')
ax3.plot(x_seg1, y_seg1, color='tab:green', lw=2,            label='avec subvention plafonnée')
ax3.plot(x_seg2, y_seg2, color='tab:green', lw=2)

ax3.axvline(x=x_bar_s, color='gray', lw=0.8, ls=':', zorder=0)
ax3.scatter([0],        [R / p2],  color='tab:blue',  zorder=4, s=30)
ax3.scatter([R / p1],   [0],       color='tab:blue',  alpha=0.5, zorder=4, s=30)
ax3.scatter([x_bar_s],  [x2_kink], color='tab:green', zorder=5, s=40)
ax3.scatter([x1_max_s], [0],       color='tab:green', zorder=4, s=30)

ax3.text(x_bar_s + 0.1, 0.22, r'$\bar x_1 = 3$', fontsize=9.5, color='gray')
ax3.text(x_bar_s + 0.12, x2_kink + 0.25,
         f'coude ({x_bar_s}, {x2_kink})', fontsize=8, color='tab:green')

ax3.set_xlim(0, 9); ax3.set_ylim(0, 8.5)
ax3.set_xlabel('$x_1$'); ax3.set_ylabel('$x_2$')
ax3.set_title(f'Subvention plafonnée ($s={s_sub}$, $\\bar x_1={x_bar_s}$)', fontsize=10)
ax3.legend(fontsize=8, loc='upper right')
ax3.grid(True, alpha=0.25)

fig.suptitle(
    r'§4.5 — Contraintes budgétaires non-linéaires'
    '\n'
    r'(référence : $R=12$, $p_1=2$, $p_2=2$  →  intercepts $(6,\ 6)$, pente $-1$)',
    fontsize=11, fontweight='bold'
)
plt.tight_layout(rect=[0, 0, 1, 0.93])
save_fig('04_5_contraintes_nonlineaires')
```
