# Chapitre 11 — Arbitrage travail-loisir

## Question du chapitre

Comment un individu arbitre-t-il entre temps travaillé, temps libre et consommation ?

## Idée à comprendre d'abord

Le loisir est un bien particulier : pour en consommer davantage, il faut renoncer à du revenu salarial. Le salaire joue donc le rôle de **coût d'opportunité du loisir**.

**Même méthodologie**  
La façon d'aborder le problème est la même que dans les chapitres précédents : on étudie un arbitrage entre deux objets, on écrit la contrainte qui borne les choix possibles, on caractérise l'optimum d'utilité sous cette contrainte, puis on analyse la réaction du comportement lorsque les conditions changent. Ici, ces deux objets ne sont plus deux biens ordinaires, mais le loisir et la consommation.

| Étape du raisonnement | Chapitres 1–10 | Chapitre 11 |
|---|---|---|
| Objet de l'arbitrage | **Ch. 1–2** : choix entre deux biens $x_1$ et $x_2$ | choix entre loisir $L$ et consommation $C$ |
| Contrainte | **Ch. 4** : budget monétaire $p_1 x_1 + p_2 x_2 = R$ | contrainte temps-revenu $C = R_0 + w(T-L)$ |
| Condition d'optimum | **Ch. 5** : $TMS = p_1/p_2$ | $TMS_{L,C} = w$ |
| Variations des conditions | **Ch. 8** : variation d'un prix ou du revenu, puis décomposition en effet substitution et effet revenu | variation du salaire ou du revenu non salarial, puis décomposition en effet substitution et effet revenu (§ 11.4) |
| Lecture économique | **Ch. 6–9** : demande de biens, élasticités, classification des biens | demande de loisir et offre de travail ; possibilité d'offre de travail rétro-courbée si le loisir est normal |
| Nature du modèle | arbitrage **sans dotation** : le revenu monétaire $R$ est donné d'avance | arbitrage **avec dotation** : l'individu dispose d'un temps total $T$ et éventuellement d'un revenu non salarial $R_0$ |

**Autre modèle de problème**  
Dans les chapitres précédents, on travaillait avec les grandeurs suivantes : $x_1$ et $x_2$ désignaient les quantités de biens consommées, $p_1$ et $p_2$ leurs prix, et $R$ un revenu monétaire donné d'avance. Le consommateur arbitrait alors entre des biens à partir d'une contrainte budgétaire :

$$
R = p_1x_1 + p_2x_2
$$

Ici, les grandeurs changent : $L$ désigne le loisir, $C$ la consommation, $T$ le temps total disponible, $w$ le salaire horaire et $R_0$ le revenu non salarial. On passe donc d'un arbitrage **sans dotation**, où le revenu est exogène, à un arbitrage **avec dotation**, où l'individu dispose d'abord d'un capital initial, ici, un capital de temps.

La contrainte qui s'écrivait auparavant  

$$
R = p_1x_1 + p_2x_2
$$  

s'écrit maintenant :

$$
R_0 + wT = C + wL 
$$

La dotation introduite ici est le temps total $T$ : c'est la ressource initiale dont dispose l'individu avant tout choix. Le terme $wT$ en est la valeur monétaire au salaire $w$, et $R_0 + wT$ représente donc les ressources initiales totales. Cela explique pourquoi une variation du salaire ne se lit pas exactement comme une simple variation de prix dans les chapitres précédents : elle modifie à la fois le coût d'opportunité du loisir et la valeur monétaire de la dotation en temps. La façon de travailler avec les graphiques pour anticiper les comportements change donc elle aussi : auparavant, on raisonnait surtout en termes de rotation ou de translation de la droite de budget ; ici, il faut aussi raisonner à partir de la dotation initiale.

## 11.1. Cadre de base

L'individu dispose d'un temps total $T$ réparti entre loisir $L$ et travail $h$ :

$$
T=L+h
$$

Avec un salaire horaire $w$ et un revenu non salarial $R_0$, la contrainte budgétaire s'écrit alors :

$$
C=R_0+w(T-L)
$$

soit :

$$
C=(R_0+wT)-wL
$$

## 11.2. Contrainte de budget travail-loisir

Dans le plan $(L,C)$, la pente de la droite de contrainte budgétaire vaut :

$$
-w
$$

Plus le salaire est élevé, plus une heure de loisir coûte cher en consommation renoncée.

## 11.3. Choix optimal

L'individu maximise :

$$
U(C,L)
$$

Quand l'optimum est intérieur, l'individu ne choisit ni le loisir complet ni le travail complet : il combine donc un peu de loisir et un peu de travail. Dans ce cas, la courbe d'indifférence est tangente à la droite de contrainte budgétaire. On a alors :

$$
TMS_{L,C}=w
$$

Le taux auquel l'individu est prêt à échanger loisir et consommation s'aligne sur le salaire du marché.

> **Méthode directe.** On peut réécrire le problème en remplaçant $C = R_0 + w(T-L)$ dans la fonction d'utilité, puis maximiser $U(R_0 + w(T-L), L)$ sans contrainte. La condition du premier ordre donne directement $-w \cdot U_C' + U_L' = 0$, soit $U_L'/U_C' = w$, c'est-à-dire $TMS = w$. Cette substitution directe est souvent plus rapide que le Lagrangien en exercice, notamment quand la forme fonctionnelle est simple. Le Lagrangien reste utile pour lire la valeur de $\lambda$ : ici $\lambda = U_C'$, l'utilité marginale de la consommation, qui est aussi l'utilité marginale du revenu.

## 11.4. Hausse du salaire

Une hausse de $w$ produit deux effets :

1. **effet substitution** : le loisir devient plus cher, donc on travaille davantage ;
2. **effet revenu** : si le loisir est un bien normal, le revenu plus élevé pousse à vouloir plus de loisir.

L'effet total sur l'offre de travail est donc ambigu.

## 11.5. Offre individuelle de travail et forme rétro-courbée

Une hausse du salaire produit à la fois un **effet substitution** et un **effet revenu**.

À bas salaire, l'effet substitution domine souvent : l'individu est incité à travailler davantage. À haut salaire, l'effet revenu peut devenir dominant : l'individu choisit davantage de loisir et réduit son offre de travail.

L'offre de travail peut donc devenir **rétro-courbée**.

La **courbe d'offre individuelle de travail** se construit en faisant varier le salaire $w$ et en relevant, pour chaque niveau, la quantité optimale de travail :

$$
h^*(w)=T-L^*(w).
$$

On obtient ainsi, dans le plan $(h,w)$, une courbe d'abord croissante lorsque l'effet substitution domine, puis éventuellement décroissante lorsque l'effet revenu domine. Le graphique plus bas illustre cette construction et fait apparaître le salaire de retournement $w^*$.

## 11.6. Salaire de réserve

Le **salaire de réserve** est le salaire minimal à partir duquel l'individu accepte de travailler.

Si le salaire de marché est inférieur à ce seuil, l'individu choisit le loisir complet. S'il lui est supérieur, il offre une quantité positive de travail.

## 11.7. Élasticité de l'offre de travail

L'élasticité de l'offre de travail mesure la sensibilité du nombre d'heures offertes à une variation du salaire :

$$
\varepsilon_{h,w} = \frac{\partial h}{\partial w} \cdot \frac{w}{h}
$$

- $\varepsilon_{h,w} > 0$ : offre croissante (effet substitution domine) ;
- $\varepsilon_{h,w} < 0$ : offre décroissante (effet revenu domine) ;
- $\varepsilon_{h,w} \approx 0$ : offre rigide, peu sensible au salaire.

Les principaux facteurs qui influencent l'élasticité sont : le niveau du revenu non salarial $R_0$, l'intensité de la préférence pour le loisir, les contraintes institutionnelles (temps partiel, conventions collectives, fiscalité).

**Élasticité de la demande de loisir.** L'élasticité de la **demande de loisir** au salaire réel s'écrit :

$$
\varepsilon_{L,w} = \frac{\partial L}{\partial w} \cdot \frac{w}{L}
$$

Elle est reliée à l'élasticité de l'offre de travail par :

$$
\varepsilon_{h,w} = -\frac{L}{h} \cdot \varepsilon_{L,w}
$$

Quand l'offre de travail est rétro-courbée ($\varepsilon_{h,w} < 0$), la demande de loisir est croissante en $w$ ($\varepsilon_{L,w} > 0$) : l'effet revenu domine et l'individu "achète" davantage de loisir avec son surcroît de richesse.

## 11.8. Boîte d'Edgeworth appliquée au ménage

Dans un ménage à deux membres, la répartition du temps et des ressources peut se représenter dans une boîte d'Edgeworth. Chaque membre a ses propres préférences entre loisir et consommation. La boîte fait apparaître :

- les allocations efficaces au sens de Pareto — la **courbe des contrats** — le long desquelles on ne peut améliorer le bien-être d'un membre sans réduire celui de l'autre ;
- la zone Pareto-supérieure aux dotations initiales ;
- les possibilités de négociation et de transferts intra-ménage.

**Interprétation économique.** L'efficacité intra-familiale est atteinte lorsqu'on se place sur la courbe des contrats. Elle dépend des salaires respectifs, des préférences et de la technologie de production domestique.

**La courbe des contrats comme continuum d'optima.** La courbe des contrats n'est pas un point mais un **continuum** d'allocations efficaces. Il existe une infinité d'optima Pareto, du coin de A au coin de B. L'efficacité au sens de Pareto ne sélectionne pas un seul optimum — elle délimite un ensemble. C'est le pouvoir de négociation, les droits de propriété ou les règles institutionnelles qui déterminent lequel est effectivement atteint.

**Salaire réel et répartition de la valeur ajoutée.** Dans la boîte d'Edgeworth, le taux de salaire réel n'est pas seulement un coût d'opportunité du loisir — il détermine aussi comment la valeur ajoutée produite est répartie entre les membres (ou entre travail et capital dans une lecture macro). Un salaire réel plus élevé déplace la solution sur la courbe des contrats en faveur du travailleur. L'enjeu de la négociation salariale est donc précisément de choisir *où* on se place sur cette courbe, parmi l'**infinité d'optima** qu'elle contient. Le modèle ne tranche pas : il montre l'espace des possibles efficaces, pas le point effectivement retenu.

## 11.9. Économie de la famille

**Becker (1965) : allocation du temps.** Chaque membre du ménage alloue son temps entre travail marchand, production domestique et loisir. L'arbitrage dépend des avantages comparatifs de chaque membre, du niveau des salaires et de la productivité domestique.

**Production domestique.** Certaines activités (cuisine, garde d'enfants, entretien) produisent des biens domestiques qui substituent ou complètent la consommation marchande. L'arbitrage entre internalisation et recours au marché dépend des prix relatifs.

**Spécialisation intra-ménage.** Si un membre a un salaire plus élevé, il peut être efficace qu'il se spécialise dans le travail marchand pendant que l'autre se spécialise dans la production domestique — logique des avantages comparatifs appliquée au foyer.

**Gronau (1977).** Distingue trois usages du temps : travail marchand, travail domestique et loisir pur. Contrairement à Becker, le loisir n'entre pas dans la fonction de production domestique ; il s'agit d'un usage direct du temps par l'individu.

**Limites.** Ces modèles supposent une décision coopérative et une information partagée. Ils ne rendent pas compte des conflits d'intérêt, des inégalités de pouvoir intra-ménage ni des dynamiques de négociation.

## Perspectives critiques et histoire de la pensée

**Robinson Crusoé chez les économistes.** Le modèle travail-loisir suppose un agent isolé qui maximise sous contrainte de temps — c'est précisément la figure de Robinson Crusoé, mobilisée par les économistes classiques pour illustrer le comportement rationnel. Marx s'en moque dans *Le Capital* (1867) : Robinson est une fiction commode qui efface les relations sociales de production en les réduisant à un calcul individuel. Edgeworth reprend lui aussi Robinson dans *Mathematical Psychics* (1881), mais pour l'opposer à un second individu — c'est l'origine de la boîte — et montrer que dès qu'il y a deux agents, les questions de répartition et de négociation deviennent inévitables.

**Gorz : le loisir colonisé.** Dans *Métamorphoses du travail* (1988), André Gorz soutient que dans les sociétés marchandes, le loisir lui-même est progressivement colonisé par la logique productive : il se transforme en temps de consommation, de récupération ou de formation, perdant son caractère de fin en soi. Le modèle néoclassique traite le loisir comme un bien parmi d'autres, soumis au même calcul d'utilité — ce que Gorz lit comme une réduction du temps libre à un temps socialement utile pour la reproduction de la force de travail. Cette critique ne réfute pas le modèle, mais en délimite la portée : il décrit un comportement d'arbitrage, pas la valeur du loisir comme expérience vécue.

## Graphique — Arbitrage travail-loisir

**Ce qu'il faut voir.** Ici, le loisir est sur l'axe horizontal et la consommation sur l'axe vertical. Quand le salaire augmente, la contrainte devient plus pentue et pivote autour du point $(T,R_0)$, car le loisir coûte davantage en consommation renoncée tandis que le loisir complet laisse toujours la consommation égale à $R_0$.

**Cas concret simple.** Si ton salaire horaire passe de 10 à 16, une heure de loisir coûte désormais 16 de consommation potentielle au lieu de 10. En revanche, si tu ne travailles pas du tout, ta consommation reste égale au revenu non salarial $R_0$ : c'est ce qui explique le pivot du graphique.

**Comment le lire vite.** Repère d'abord le point fixe $(T,R_0)$, puis compare la pente des deux droites budgétaires. Ce graphique montre donc surtout comment la contrainte change quand le salaire augmente ; il ne montre pas encore, à lui seul, la décomposition entre effet substitution et effet revenu.

**Point concours.** Ce graphique montre bien le pivot de la contrainte, mais pas encore la séparation entre effet substitution et effet revenu. Pour aller plus loin, il faudrait appliquer une des méthodes de décomposition présentées au chapitre 8.

```python
# 9 labor leisure

T = 24

w1 = 1

w2 = 1.6

R0 = 4  # revenu non salarial

L = np.linspace(0, T, 200)

pivot = (T, R0)

budget_initial = R0 + w1 * (T - L)

budget_high = R0 + w2 * (T - L)

plt.figure(figsize=(6.4, 4.8))

plt.plot(L, budget_initial, label='salaire initial', color='tab:blue', linewidth=2)

plt.plot(L, budget_high, label='salaire plus élevé', color='tab:orange', linewidth=2)

# Le point fixe est le loisir complet : si L = T, on ne travaille pas et C = R0 quel que soit w.

plt.scatter(*pivot, color='black', s=40, zorder=5)

plt.annotate('pivot : loisir complet\n$(T, R_0)$', xy=pivot, xytext=(3.6, 7.0),

             arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)

plt.text(13.2, 38.2, 'Contrainte : $C = (R_0 + wT) - wL $\navec $R_0$ = revenu non salarial',
         fontsize=9, bbox=dict(boxstyle='round,pad=0.35', facecolor='white', edgecolor='0.4', alpha=0.95))

plt.xlabel('loisir L (heures)')

plt.ylabel('consommation C (€)')

plt.title('Arbitrage travail-loisir')

plt.xlim(0, 26)

plt.ylim(0, 45)

plt.grid(True, alpha=.25)

plt.legend()

save_fig('09_arbitrage_travail_loisir')
```

### Offre de travail rétro-courbée

Le graphique représente l'offre individuelle de travail $h^*(w)$ en fonction du salaire $w$. À bas salaire, l'**effet de substitution** domine et l'offre croît. Au-delà du salaire de retournement $w^*$, l'**effet revenu** domine et l'offre diminue : on parle d'**offre rétro-courbée** (*backward bending*). L'élasticité de l'offre passe de positive à négative en ce point.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

FIG_DIR = Path("microeco_v5_figures")
FIG_DIR.mkdir(exist_ok=True)

def save_fig(name):
    plt.tight_layout()
    plt.savefig(FIG_DIR / f"{name}.png", dpi=150, bbox_inches="tight")
    plt.savefig(FIG_DIR / f"{name}.svg", bbox_inches="tight")
    plt.show()

fig, ax = plt.subplots(figsize=(5.5, 4.5))

# Offre rétro-courbée : h = 2 + 0.8w - 0.025w²  (pic en w*=16)
w = np.linspace(0, 26, 300)
h = 2 + 0.8 * w - 0.025 * w**2
h = np.clip(h, 0, None)

ax.plot(h, w, 'b-', lw=2.5, label="Offre de travail $h^*(w)$")

w_star, h_star = 16, 8.4
ax.plot(h_star, w_star, 'ko', ms=7, zorder=5)
ax.annotate("$w^*$ — retournement",
            xy=(h_star, w_star), xytext=(h_star - 3.8, w_star + 3.5),
            fontsize=8.5, ha='left',
            arrowprops=dict(arrowstyle='->', color='gray', lw=1.2))

ax.annotate("effet substitution\ndomine",  xy=(5.0, 7),  fontsize=8, color='navy',   style='italic')
ax.annotate("effet revenu\ndomine",        xy=(6.5, 21), fontsize=8, color='darkred', style='italic')

ax.set_xlabel("Heures travaillées $h$", fontsize=10)
ax.set_ylabel("Salaire $w$", fontsize=10)
ax.set_title("Offre de travail rétro-courbée (*backward bending*)", fontsize=10.5)
ax.set_xlim(0, 11); ax.set_ylim(0, 28)

box_text = (r"$h^*(w) = 2 + 0{,}8w - 0{,}025w^2$" + "\n"
            r"Pic en $w^* = 16$, $h^* \approx 8{,}4$" + "\n"
            r"$\varepsilon_{h,w} = \frac{\partial h}{\partial w}\cdot\frac{w}{h}$")
ax.text(0.97, 0.05, box_text, transform=ax.transAxes,
        fontsize=8.0, ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.35', facecolor='white', edgecolor='0.4', alpha=0.95))

ax.grid(alpha=0.3)
ax.spines[['top', 'right']].set_visible(False)
save_fig('09b_offre_travail_retro')
```

## Graphique — Boîte d'Edgeworth appliquée au ménage

**Ce qu'il faut voir.** La boîte représente l'ensemble des répartitions possibles d'un volume total de loisir et de consommation entre deux membres du ménage. Le point $E$ est une allocation initiale. Les deux courbes d'indifférence qui passent par $E$ jouent le rôle de frontières : elles séparent, pour chaque membre, les allocations au moins aussi bonnes que $E$ de celles qui lui sont inférieures. La **lentille d'amélioration mutuelle** est l'intersection de ces deux ensembles.

**Pourquoi la courbe de B est-elle "dans l'autre sens" ?** Parce qu'on représente ici les quantités du membre A sur les axes. Quand on va vers la droite ou vers le haut, A reçoit davantage, donc B reçoit moins du total disponible. Pour B, l'origine pertinente n'est donc pas en bas à gauche, mais en **haut à droite** de la boîte : sa courbe d'indifférence paraît inversée simplement parce qu'on la lit depuis son propre coin d'origine. Recevoir moins en quantité sur les axes de A ne signifie donc pas automatiquement être moins bien loti : pour B, tout dépend de la position du point par rapport à sa propre courbe d'indifférence passant par $E$.

**Comment atteint-on la courbe des contrats ?** À partir de la dotation initiale $E$, les membres du ménage ont intérêt à se réallouer temps et ressources tant qu'il existe un point dans la lentille qui améliore la situation de l'un sans détériorer celle de l'autre. Le processus s'arrête quand ces gains mutuels sont épuisés : on atteint alors un point de la **courbe des contrats**, c'est-à-dire une allocation efficace au sens de Pareto.

**Lecture économique.** La **courbe des contrats** relie donc les allocations efficaces, mais elle ne dit pas à elle seule quel point sera retenu. Le point effectivement atteint dépend du pouvoir de négociation, des salaires relatifs, des préférences et, plus largement, des règles de partage au sein du ménage.

**Pourquoi ce graphique est utile ici.** Il montre d'un seul coup quatre idées que le texte seul laisse abstraites : la dotation initiale, la possibilité de gains mutuels, le mécanisme qui conduit vers une allocation efficace, puis la distinction entre efficacité et répartition.

```python
# Edgeworth box - household allocation of leisure and consumption
L_tot = 10
C_tot = 10
alpha_a = 0.6
alpha_b = 0.4

L_a = np.linspace(0.2, L_tot - 0.2, 400)
k_a = alpha_a / (1 - alpha_a)
k_b = alpha_b / (1 - alpha_b)
C_contract = (k_b * C_tot * L_a) / (k_a * (L_tot - L_a) + k_b * L_a)

L_a_endowment = 3.2
C_a_endowment = 3.0
u_a_endowment = (L_a_endowment ** alpha_a) * (C_a_endowment ** (1 - alpha_a))
u_b_endowment = ((L_tot - L_a_endowment) ** alpha_b) * ((C_tot - C_a_endowment) ** (1 - alpha_b))

grid_l = np.linspace(0.2, L_tot - 0.2, 250)
grid_c = np.linspace(0.2, C_tot - 0.2, 250)
LL, CC = np.meshgrid(grid_l, grid_c)
UA = (LL ** alpha_a) * (CC ** (1 - alpha_a))
UB = ((L_tot - LL) ** alpha_b) * ((C_tot - CC) ** (1 - alpha_b))
improve_both = (UA >= u_a_endowment) & (UB >= u_b_endowment)

plt.figure(figsize=(6.8, 6.0))
plt.contourf(LL, CC, improve_both.astype(int), levels=[0.5, 1.5], colors=['#dff3e3'], alpha=0.9)

c_a = (u_a_endowment / (L_a ** alpha_a)) ** (1 / (1 - alpha_a))
mask_a = (c_a > 0) & (c_a < C_tot)
plt.plot(L_a[mask_a], c_a[mask_a], color='tab:blue', linewidth=2, label="indifférence de A")

L_for_b = np.linspace(0.2, L_tot - 0.2, 400)
c_b = C_tot - (u_b_endowment / ((L_tot - L_for_b) ** alpha_b)) ** (1 / (1 - alpha_b))
mask_b = (c_b > 0) & (c_b < C_tot)
plt.plot(L_for_b[mask_b], c_b[mask_b], color='tab:red', linewidth=2, label="indifférence de B")

plt.plot(L_a, C_contract, color='black', linestyle='--', linewidth=2, label='courbe des contrats')
plt.scatter([L_a_endowment], [C_a_endowment], color='black', s=45, zorder=5)
plt.annotate('E : allocation initiale', xy=(L_a_endowment, C_a_endowment), xytext=(1.2, 1.4),
             arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)

plt.text(6.0, 8.6, 'zone d\'amélioration\nmutuelle', fontsize=9, color='darkgreen',
         bbox=dict(boxstyle='round,pad=0.25', facecolor='white', edgecolor='darkgreen', alpha=0.9))

plt.xlim(0, L_tot)
plt.ylim(0, C_tot)
plt.xlabel('loisir du membre A')
plt.ylabel('consommation du membre A')
plt.title("Boîte d'Edgeworth : gains mutuels et efficacité dans le ménage")
plt.grid(True, alpha=.2)
plt.legend(loc='lower right', fontsize=8)
save_fig('09c_edgeworth_menage')
```

## Graphique — Becker : production domestique en deux étapes

**Idée générale.** Chez Becker, le ménage produit un service domestique $Z$ en combinant des achats marchands $x$ et du temps domestique $t_d$, puis arbitre entre ce service et le loisir $L$.

**Variables.** $Z$ désigne le niveau de service domestique ; $x$ les achats marchands ; $t_d$ le temps domestique ; $L$ le loisir ; $p$ le prix des achats marchands ; $w$ le coût d'opportunité du temps ; $T$ le temps total disponible.

**Objectif.** Le problème final est :

$$
\max_{Z,L} \; U(Z,L) \quad \text{sous la contrainte} \quad \pi_Z Z + wL = wT.
$$

Ici, $\pi_Z$ est le coût minimal de production d'une unité de service domestique. Pour l'obtenir, on résout d'abord :

$$
\min_{x,t_d} \; px + wt_d \quad \text{sous la contrainte} \quad f(x,t_d)=Z.
$$

**Exemple retenu.** Pour le graphique, on prend une forme Cobb-Douglas simple. Plus rigoureusement, on peut écrire $Z=A\sqrt{x\,t_d}$, puis normaliser $A=1$ pour alléger le calcul et le graphique. Cette forme traduit l'idée que les deux intrants sont nécessaires et que, si l'un reste fixé, augmenter l'autre aide bien à produire davantage mais de moins en moins fortement.

Avec cette spécification, on obtient directement :

$$
c(w,p,Z)=2Z\sqrt{wp}, \qquad \pi_Z = \frac{\partial c(w,p,Z)}{\partial Z} = 2\sqrt{wp}.
$$

**Lecture géométrique.** La géométrie n'est pas nécessaire pour le calcul, mais elle permet de visualiser le coût minimal. Pour un niveau donné $Z_0$, l'**isoquante** est la courbe de toutes les combinaisons $(x,t_d)$ qui permettent de produire exactement ce même niveau $Z_0$. Ici, elle est donnée par :

$$
Z_0 = \sqrt{x t_d} \quad \Longrightarrow \quad t_d = \frac{Z_0^2}{x},
$$

et une droite d'isocoût vérifie :

$$
t_d = \frac{\bar C}{w} - \frac{p}{w}x.
$$

Le point de tangence entre l'isoquante et l'isocoût le plus bas donne la combinaison $(x,t_d)$ la moins coûteuse pour produire $Z_0$.

**Lecture du schéma.** Le panneau de gauche compare un salaire initial et un salaire plus élevé dans le problème de coût minimal. Le panneau de droite montre comment cette hausse modifie ensuite la contrainte en revenu complet et déplace l'optimum entre service domestique et loisir.

**Lien avec le salaire de réserve.** Chez Becker, le salaire de réserve n'est pas exogène : il dépend du prix implicite $\pi_Z$. Si le remplacement marchand de la production domestique est coûteux, alors $\pi_Z$ est élevé et il faut un salaire de marché plus fort pour rendre l'activité marchande attractive. C'est l'intuition souvent mobilisée empiriquement pour les ménages avec jeunes enfants : si la garde d'enfants est chère, le salaire de réserve est plus élevé et la participation au marché est plus faible.

**Effet d'une hausse du salaire.** Une hausse de $w$ est ambiguë ici aussi. D'un côté, elle renchérit le temps domestique et pousse à substituer des achats marchands au temps domestique. De l'autre, elle augmente aussi $\pi_Z$, donc renchérit le service domestique lui-même dans le choix final entre $Z$ et $L$. L'effet net sur le loisir $L$ et sur le travail marchand dépend donc à la fois des préférences et de la technologie domestique, ce qui rend Becker plus riche que le modèle travail-loisir de base.

**Corollaire intra-ménage.** Si deux membres du ménage ont des salaires différents, l'efficacité pousse celui qui a le salaire le plus élevé à se spécialiser davantage dans le travail marchand, tandis que l'autre se spécialise davantage dans la production domestique. On retrouve ici la logique des avantages comparatifs, et c'est le pont naturel entre Becker et la boîte d'Edgeworth : la courbe des contrats peut se lire comme l'ensemble des répartitions efficaces de ces spécialisations au sein du ménage.

**Idée économique.** Becker enrichit donc le modèle standard sur trois plans : il fait dépendre le salaire de réserve de la technologie domestique, il rend l'effet d'une hausse du salaire plus complexe que dans le cadre simple, et il fournit une base microéconomique à la spécialisation intra-ménage.

**À savoir faire.**

- expliquer que $w$ et $p$ déterminent d'abord le coût domestique $c(w,p,Z)$, puis le prix implicite $\pi_Z$ ;
- relier $\pi_Z$ au salaire de réserve et à la participation au marché ;
- commenter l'effet ambigu d'une hausse de $w$ sur $Z$, $L$ et le travail marchand ;
- faire le lien entre avantages comparatifs, spécialisation intra-ménage et boîte d'Edgeworth.

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    save_fig
except NameError:
    from pathlib import Path
    FIG_DIR = Path("microeco_v5_figures")
    FIG_DIR.mkdir(exist_ok=True)
    def save_fig(name):
        plt.tight_layout()
        plt.savefig(FIG_DIR / f"{name}.png", bbox_inches="tight")
        plt.savefig(FIG_DIR / f"{name}.svg", bbox_inches="tight")
        plt.show()

# Becker : effet d'une hausse du salaire
z0 = 4.0
p_market = 1.0
w_low = 2.0
w_high = 4.0
x_market = np.linspace(0.8, 11.0, 500)
t_domestic = (z0 ** 2) / x_market
mask_iso = (t_domestic > 0) & (t_domestic < 10)

def cost_minimizer(z_level, wage, price):
    x_star = z_level * np.sqrt(wage / price)
    t_star = z_level * np.sqrt(price / wage)
    cost_star = price * x_star + wage * t_star
    return x_star, t_star, cost_star

x_star_low, t_star_low, cost_low = cost_minimizer(z0, w_low, p_market)
x_star_high, t_star_high, cost_high = cost_minimizer(z0, w_high, p_market)
isocost_low = (cost_low / w_low) - (p_market / w_low) * x_market
isocost_high = (cost_high / w_high) - (p_market / w_high) * x_market
mask_cost_low = (isocost_low > 0) & (isocost_low < 10)
mask_cost_high = (isocost_high > 0) & (isocost_high < 10)

T_choice = 16.0
a_service = 0.9
b_leisure = 1.1

def budget_and_optimum(wage, price, total_time):
    pi_z = 2 * np.sqrt(wage * price)
    z_max = wage * total_time / pi_z
    z_grid = np.linspace(0.05, z_max - 0.05, 500)
    leisure = (wage * total_time - pi_z * z_grid) / wage
    utility = a_service * np.sqrt(z_grid) + b_leisure * np.sqrt(leisure)
    index_star = np.argmax(utility)
    return {
        "pi_z": pi_z,
        "z_grid": z_grid,
        "leisure": leisure,
        "z_star": z_grid[index_star],
        "l_star": leisure[index_star],
        "u_star": utility[index_star],
        "z_max": z_max,
    }

scenario_low = budget_and_optimum(w_low, p_market, T_choice)
scenario_high = budget_and_optimum(w_high, p_market, T_choice)
z_curve = np.linspace(0.05, max(scenario_low["z_max"], scenario_high["z_max"]) - 0.05, 500)

def indifference_curve(u_level, z_values):
    residual = u_level - a_service * np.sqrt(z_values)
    mask = residual > 0
    leisure = np.full_like(z_values, np.nan)
    leisure[mask] = (residual[mask] / b_leisure) ** 2
    return leisure

indiff_low = indifference_curve(scenario_low["u_star"], z_curve)
indiff_high = indifference_curve(scenario_high["u_star"], z_curve)
mask_indiff_low = (indiff_low > 0) & (indiff_low < T_choice)
mask_indiff_high = (indiff_high > 0) & (indiff_high < T_choice)

fig, axes = plt.subplots(1, 2, figsize=(11.8, 5.0))

ax = axes[0]
ax.plot(x_market[mask_iso], t_domestic[mask_iso], color='black', linewidth=2.5, label='isoquante $Z_0$')
ax.plot(x_market[mask_cost_low], isocost_low[mask_cost_low], color='tab:blue', linestyle='--', linewidth=2, label='isocoût : salaire initial')
ax.plot(x_market[mask_cost_high], isocost_high[mask_cost_high], color='tab:orange', linestyle='--', linewidth=2, label='isocoût : salaire plus élevé')
ax.scatter([x_star_low, x_star_high], [t_star_low, t_star_high], color=['tab:blue', 'tab:orange'], s=42, zorder=5)
ax.annotate('optimum initial', xy=(x_star_low, t_star_low), xytext=(3.5, 4.2),
            arrowprops=dict(arrowstyle='->', color='tab:blue'), fontsize=9, color='tab:blue')
ax.annotate('nouvel optimum', xy=(x_star_high, t_star_high), xytext=(8.2, 2.8),
            arrowprops=dict(arrowstyle='->', color='tab:orange'), fontsize=9, color='tab:orange')
ax.text(0.05, 0.08, r'$Z = \sqrt{x\,t_d}$' + '\n' + r'$\min px + wt_d$', transform=ax.transAxes, fontsize=9,
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='0.4', alpha=0.95))
ax.set_xlabel('achats marchands $x$')
ax.set_ylabel('temps domestique $t_d$')
ax.set_title('1. Hausse du salaire et coût de production')
ax.set_xlim(0, 11)
ax.set_ylim(0, 10)
ax.grid(True, alpha=.25)
ax.legend(loc='upper right', fontsize=8)

ax = axes[1]
ax.plot(scenario_low["z_grid"], scenario_low["leisure"], color='tab:blue', linestyle='--', linewidth=2.2, label='budget : salaire initial')
ax.plot(scenario_high["z_grid"], scenario_high["leisure"], color='tab:orange', linestyle='--', linewidth=2.2, label='budget : salaire plus élevé')
ax.plot(z_curve[mask_indiff_low], indiff_low[mask_indiff_low], color='tab:blue', linewidth=2, alpha=0.9, label='indifférence : optimum initial')
ax.plot(z_curve[mask_indiff_high], indiff_high[mask_indiff_high], color='tab:orange', linewidth=2, alpha=0.9, label='indifférence : nouvel optimum')
ax.scatter([scenario_low["z_star"], scenario_high["z_star"]], [scenario_low["l_star"], scenario_high["l_star"]],
           color=['tab:blue', 'tab:orange'], s=42, zorder=5)
ax.annotate('optimum initial', xy=(scenario_low["z_star"], scenario_low["l_star"]), xytext=(3.1, 12.2),
            arrowprops=dict(arrowstyle='->', color='tab:blue'), fontsize=9, color='tab:blue')
ax.annotate('nouvel optimum', xy=(scenario_high["z_star"], scenario_high["l_star"]), xytext=(8.4, 10.2),
            arrowprops=dict(arrowstyle='->', color='tab:orange'), fontsize=9, color='tab:orange')
ax.text(0.05, 0.08, r'$\pi_Z = 2\sqrt{wp}$' + '\n' + r'$\max U(Z,L)$', transform=ax.transAxes, fontsize=9,
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='0.4', alpha=0.95))
ax.set_xlabel('service domestique $Z$')
ax.set_ylabel('loisir $L$')
ax.set_title('2. Hausse du salaire et choix final')
ax.set_xlim(0, max(scenario_low["z_max"], scenario_high["z_max"]) + 0.4)
ax.set_ylim(0, T_choice)
ax.grid(True, alpha=.25)
ax.legend(loc='upper right', fontsize=8)

fig.suptitle('Becker : effet d\'une hausse du salaire', fontsize=12, y=1.02)
save_fig('11a_becker_deux_etapes')
```

## Graphique — Gronau : trois usages du temps

**Idée générale.** Chez Gronau, le temps total se répartit entre travail marchand, travail domestique et loisir pur. Le point important est que le loisir n'est plus traité comme un intrant de la production domestique : c'est un usage direct du temps.

**Variables.** $T$ désigne le temps total disponible ; $h_m$ le temps de travail marchand ; $h_d$ le temps de travail domestique ; $L_e$ le loisir ; $Z$ la production domestique ; $x$ la consommation achetée sur le marché ; $w$ le salaire.

**Objectif.** Le premier panneau visualise la contrainte de temps entre trois usages concurrents. Le second montre le graphe canonique de Gronau dans l'espace $(L_e, Z+x)$, là où se lit l'arbitrage économique.

**1. Contrainte de temps.** Le point de départ est :

$$
T = h_m + h_d + L_e.
$$

Si l'on fixe un niveau de loisir $L_e=L_0$, on obtient dans le plan $(h_m,h_d)$ :

$$
h_d = T - L_0 - h_m.
$$

Toutes les droites de loisir constant ont donc une pente $-1$ et sont parallèles. La frontière supérieure correspond au cas $L_e=0$, soit $h_d = T - h_m$.

**2. Graphe canonique dans $(L_e, Z+x)$.** Si tout le temps hors loisir est d'abord affecté au travail domestique, on obtient un arc concave de production domestique. Au point où la valeur du produit marginal du travail domestique est égale au salaire, soit $pMg_H = w$, cet arc devient tangent à une droite salariale de pente $-w$. Au-delà de ce point de bascule, une heure supplémentaire hors loisir est affectée au marché plutôt qu'à la maison.

**Condition d'entrée sur le marché du travail.** Un individu ne travaille sur le marché que si son salaire dépasse la valeur marginale de sa production domestique. Tant que $w < pMg_H$, il reste en production domestique. Cette condition détermine le salaire de réserve dans le modèle et relie directement Gronau au salaire de réserve étudié au §11.6.

**Effet d'une hausse du salaire.** Quand $w$ augmente, le point de tangence se déplace le long de l'arc : le travail domestique diminue et le travail marchand augmente. En revanche, l'effet final sur le loisir reste ambigu, pour la même raison que dans le modèle de base : il dépend de l'arbitrage entre effet substitution et effet revenu.

**Lecture du schéma.** Le panneau de gauche montre le triangle des allocations possibles. Le panneau de droite montre le graphique canonique : l'arc de production domestique, les droites salariales tangentes et les points de bascule où $pMg_H = w$. Le point $A$ du panneau de gauche est seulement un exemple d'allocation : il ne représente pas, à lui seul, un optimum.

**Idée économique.** Par rapport à Becker, Gronau isole davantage le loisir et fait apparaître la condition de passage du travail domestique au travail marchand. C'est cette distinction qui permet d'analyser correctement l'entrée sur le marché du travail.

- tracer le graphe canonique dans $(L_e, Z+x)$ avec l'arc domestique puis la droite salariale tangente ;
- localiser le point de bascule où $pMg_H = w$ ;
- formuler la condition d'entrée sur le marché du travail et l'interpréter comme un salaire de réserve ;
- expliquer qu'une hausse de $w$ réduit $h_d$, augmente $h_m$ et laisse l'effet sur le loisir indéterminé.

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    save_fig
except NameError:
    from pathlib import Path
    FIG_DIR = Path("microeco_v5_figures")
    FIG_DIR.mkdir(exist_ok=True)
    def save_fig(name):
        plt.tight_layout()
        plt.savefig(FIG_DIR / f"{name}.png", bbox_inches="tight")
        plt.savefig(FIG_DIR / f"{name}.svg", bbox_inches="tight")
        plt.show()

# Gronau : triangle des usages du temps et graphe canonique
T_total = 16
h_market = np.linspace(0, T_total, 400)
h_home_frontier = T_total - h_market

def gronau_switch(wage, productivity):
    h_home_star = min((productivity / (2 * wage)) ** 2, T_total)
    leisure_switch = T_total - h_home_star
    output_switch = productivity * np.sqrt(h_home_star)
    return h_home_star, leisure_switch, output_switch

A_home = 6.0
w_low = 0.9
w_high = 1.4
leisure_grid = np.linspace(0, T_total, 500)
home_arc = A_home * np.sqrt(np.maximum(T_total - leisure_grid, 0))

h_home_low, leisure_switch_low, output_switch_low = gronau_switch(w_low, A_home)
h_home_high, leisure_switch_high, output_switch_high = gronau_switch(w_high, A_home)

line_low = output_switch_low + w_low * (leisure_switch_low - leisure_grid)
line_high = output_switch_high + w_high * (leisure_switch_high - leisure_grid)
frontier_low = np.where(leisure_grid <= leisure_switch_low, line_low, home_arc)
frontier_high = np.where(leisure_grid <= leisure_switch_high, line_high, home_arc)

fig, axes = plt.subplots(1, 2, figsize=(12.2, 5.2))

ax = axes[0]
ax.fill_between(h_market, 0, h_home_frontier, color='#e8f1fb', alpha=0.9)
ax.plot([0, T_total, 0, 0], [0, 0, T_total, 0], color='#8aa9c4', linewidth=1.4)
ax.plot(h_market, h_home_frontier, color='tab:blue', linewidth=2.5, label='frontière $h_m + h_d = T$')

L_level = 4
h_home_leisure = T_total - L_level - h_market
mask_leisure = h_home_leisure >= 0
ax.plot(h_market[mask_leisure], h_home_leisure[mask_leisure], color='tab:orange', linestyle='--', linewidth=2, label='loisir constant $L_e=4$')

h_m_star, h_d_star = 6, 6
ax.scatter([h_m_star], [h_d_star], color='black', s=45, zorder=5)
ax.annotate('A : exemple d\'allocation\n$L_e = 16 - 6 - 6 = 4$', xy=(h_m_star, h_d_star), xytext=(7.1, 8.3),
            arrowprops=dict(arrowstyle='->', color='black'), fontsize=9)
ax.text(3.0, 4.0, 'triangle des\nallocations\npossibles', fontsize=8.5, color='#315a7d',
        bbox=dict(boxstyle='round,pad=0.25', facecolor='white', edgecolor='#8aa9c4', alpha=0.9))
ax.text(0.54, 0.08, r'$T = h_m + h_d + L_e$', transform=ax.transAxes, fontsize=9,
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='0.4', alpha=0.95))
ax.set_xlabel('travail marchand $h_m$')
ax.set_ylabel('travail domestique $h_d$')
ax.set_title('1. Triangle des usages du temps')
ax.set_xlim(0, T_total)
ax.set_ylim(0, T_total)
ax.grid(True, alpha=.25)
ax.legend(loc='upper right', fontsize=8)
ax.spines[['top', 'right']].set_visible(False)

ax = axes[1]
ax.plot(leisure_grid, home_arc, color='black', linewidth=2.4, label='arc de production domestique')
ax.plot(leisure_grid, frontier_low, color='tab:blue', linestyle='--', linewidth=2.2, label='frontière : salaire initial')
ax.plot(leisure_grid, frontier_high, color='tab:orange', linestyle='--', linewidth=2.2, label='frontière : salaire plus élevé')
ax.scatter([leisure_switch_low, leisure_switch_high], [output_switch_low, output_switch_high], color=['tab:blue', 'tab:orange'], s=44, zorder=5)
ax.annotate(r'$pMg_H = w_0$', xy=(leisure_switch_low, output_switch_low), xytext=(3.1, 19.4),
            arrowprops=dict(arrowstyle='->', color='tab:blue'), fontsize=9, color='tab:blue')
ax.annotate(r'$pMg_H = w_1$', xy=(leisure_switch_high, output_switch_high), xytext=(10.1, 15.4),
            arrowprops=dict(arrowstyle='->', color='tab:orange'), fontsize=9, color='tab:orange')
ax.annotate('hausse de $w$', xy=(leisure_switch_high, output_switch_high + 0.6), xytext=(leisure_switch_low + 0.6, output_switch_low + 2.3),
            arrowprops=dict(arrowstyle='->', color='0.35'), fontsize=8.8, color='0.25')
ax.text(0.04, 0.08, r'entrée sur le marché : $w > pMg_H$' + '\n' + r'salaire de réserve : $w = pMg_H$',
        transform=ax.transAxes, fontsize=8.8,
        bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='0.4', alpha=0.95))
ax.set_xlabel('loisir $L_e$')
ax.set_ylabel('service total $Z + x$')
ax.set_title('2. Graphe canonique de Gronau')
ax.set_xlim(0, T_total)
ax.set_ylim(0, max(frontier_high.max(), frontier_low.max()) + 1.5)
ax.grid(True, alpha=.25)
ax.legend(loc='upper right', fontsize=8)
ax.spines[['top', 'right']].set_visible(False)

fig.suptitle('Gronau : contrainte de temps et arbitrage économique', fontsize=12, y=1.02)
save_fig('11b_gronau_trois_usages_temps')
```
